```python
from diffusion import DDMFunctions, Network, Train
```

```python
image_size = 28
timesteps=300
channels = 1
```

```python
## Unet
import torch
device = "cuda" if torch.cuda.is_available() else "cpu"

model = Network.Unet(dim=image_size, channels=channels)
_ = model.to(device)
```

# 学習

```python
dataset_name = "fashion_mnist"
batch_size = 512 
epochs=6 
save_and_sample_every = 500 
schedule_type = "linear"
```

```python
train = Train.Train(model, image_size, channels, timesteps, dataset_name, schedule_type, device)
```

```python
train.train(epochs, save_and_sample_every=save_and_sample_every, batch_size=batch_size)
```

```python
torch.save(model.state_dict(), 'fashion_mnist_modek.pkl')
```

```python
sample_batch_size = 10
dm = DDMFunctions.DDMFunctions(timesteps, schedule_type)
samples = dm.sample(model, image_size, sample_batch_size, channels)
```

```python
## 生成結果の表示

import matplotlib.pyplot as plt
random_index = 4 
plt.imshow(samples[-1][random_index].reshape(image_size, image_size, channels), cmap="gray")
```

```python
## 逆拡散プロセスのアニメーション化
gif_file_name = "/content/denoising_diffusion_process.gif" #@param{type:"string"}
import matplotlib.animation as animation

fig = plt.figure()
ims = []
for i in range(timesteps):
    im = plt.imshow(samples[i][random_index].reshape(image_size, image_size, channels), cmap="gray", animated=True)
    ims.append([im])

animate = animation.ArtistAnimation(fig, ims, interval=50, blit=True, repeat_delay=1000)
animate.save(gif_file_name)
```
