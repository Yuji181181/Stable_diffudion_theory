```python
import numpy as np
import math

pi = math.pi
exp = math.exp
sqrt = math.sqrt
```

# ガウス分布の確率密度関数

$$
X \sim \mathcal{N}(\mu,\,\sigma^{2})
$$


$$
f(x) = \frac{1}{{\sigma \sqrt{2\pi }}}e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2}
$$

```python
def f(σ, μ, x):
    係数 = 1/(σ * sqrt(2*pi))
    項 = exp(-0.5 * ((x-μ)/σ )**2 )
    return 係数 * 項
```

```python
σ = 1 # 標準偏差
μ = 0 # 平均
x = 2
f(σ, μ, x)
```

```python
from matplotlib import pyplot as plt
```

```python
# @title 確率密度関数のグラフ
σ = 1 #@param {type:"slider", min:0.1, max:2.0, step:0.1}
μ = 0 #@param {type:"slider", min:-2, max:2.0, step:0.1}
result = []
for i in range(1000):
    x = (i - 500) * 0.01
    y = f(σ, μ, x)
    result.append(y)
plt.plot(result)
# x軸の目盛設定
# plt.set_xticks([-4, -2, 0, 2, 4])
# x軸の目盛りの設定
locs, labels = plt.xticks()
new_labels = [str(float(label)/100 - 5) for label in locs]
plt.xticks(locs, new_labels)
plt.show()
```

# 積分の計算で確率を求めましょう

```python
# @markdown Initial point.
a = -10 # @param {type:"slider", min:-10.0, max:10.0, step:0.1}
# @markdown Final point.
b = 0 # @param {type:"slider", min:-10.0, max:10.0, step:0.1}
step_size = 0.0001 #@param {type:'number'}

# 積分の実行
sum = 0;
x = a
while x < b:
    y = f(σ, μ, x)
    sum += y * step_size;

    x += step_size

# 結果の表示
print("積分結果:", sum)
```
